```python
from ws_maps.route import Origin, Destination, Engine
from ws_maps.network import Network
import time

network = Network()
origin = Origin(address='2120 McGee Ave, Berkeley, CA 94703')
destination = Destination(address='161 Steuart St, San Francisco, CA 94105')
engine = Engine(network=network, traffic=True, terrain=True, weather=True)
start_time = time.time()
matched_route = engine.route(origin, destination, matched=True)
print("routing took {} seconds".format(time.time() - start_time))
matched_route.plot()
start_time = time.time()
non_matched_route = engine.route(origin, destination, matched=False)
print("routing took {} seconds".format(time.time() - start_time))
non_matched_route.plot()
```
